```python
df.info() #give datatype
df.describe() #gives statistical value for only numerical column
df.describe(include = 'all') #gives statistical info about all the columns
```

```python
df.isnull().sum() # review rating has 37 missing values-- median based on each category's review
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
df.isnull().sum() #Review rating has become null
```

```python
df.columns = df.columns.str.lower() #lower case
df.columns = df.columns.str.replace(' ','_') #replacing space with'_' for easy cleaning
df.columns
```

```python
df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
age_labels = {'Young adult','Adult','Mid-aged','Senior'}
df['age_group']=pd.qcut(df['age'],q=4,labels=age_labels)
df[['age','age_group']].head(5)
```
